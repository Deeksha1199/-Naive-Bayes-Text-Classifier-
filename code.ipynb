```python
# Average filtering with kernel size 3x3
m, n = task4_image.shape[:2] 
#Creating a kernel filter with size 3x3
kernel_filter = numpy.ones([3 , 3], dtype = int)
#finding the average for 3x3 for the kernel 
kernel_filter = kernel_filter / 9  
#padding the image   
filtered_image = numpy.zeros([m + 2, n + 2])
#applying filter
filtered_image = signal.convolve2d(task4_image , kernel_filter) 
plt.imshow(filtered_image, cmap = 'gray') 
plt.show()
```

```python
#Reading the image
task5_image = cv2.imread("./Images/cameraman.jpeg", 0)
```

```python
# Applying salt and pepper noise by 30% to this image
image = task5_image/255
empty_image = numpy.zeros((image.shape[0],image.shape[1]), dtype = numpy.float32 )
pepper = 0.3
salt = 1 - pepper
i=0
j =0
width, height = image.shape[:2]
for i in range(width):
  for j in range(height):
    N= numpy.random.random()
    if N < pepper:
      empty_image[i][j] = 0
    elif N > salt:
      empty_image[i][j] = 1
    else:
      empty_image[i][j] = image[i][j]
    j = j+1
  i = i +1

plt.imshow(empty_image , cmap = "gray")
plt.show()
```

```python
# Applying median filter with kernel size 3x3
filtered_images = []
#Creating a kernel filter with size 3x3
kernel_filter = numpy.ones([3 , 3], dtype = int)
#applying median filter
filtered_image = ndimage.median_filter(task5_image, footprint=kernel_filter)
filtered_images.append(filtered_image)
plt.imshow(filtered_image , cmap = "gray")
plt.show()
```

```python
# Applying median filter with kernel size 5x5
filtered_images = []
#Creating a kernel filter with size 5x5
kernel_filter = numpy.ones([5 , 5], dtype = int)
#applying median filter
filtered_image = ndimage.median_filter(task5_image, footprint=kernel_filter)
filtered_images.append(filtered_image)
plt.imshow(filtered_image , cmap = "gray")
plt.show()
```

```python
# Applying median filter with kernel size 9x9
filtered_images = []
#Creating a kernel filter with size 9x9
kernel_filter = numpy.ones([9 , 9], dtype = int)
#applying median filter
filtered_image = ndimage.median_filter(task5_image, footprint=kernel_filter)
filtered_images.append(filtered_image)
plt.imshow(filtered_image , cmap = "gray")
plt.show()
```

```python
#reading the image
task6_image = cv2.imread('./Images/cameraman.jpeg', 0)
```

```python
# Sobel Horizontal Filter
horizontal_sobel = numpy.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]]) 
#convoluting the filter with the image
filtered_image_horizontal =  signal.convolve2d(task6_image , horizontal_sobel)
plt.imshow(filtered_image_horizontal , cmap = "gray")
plt.show()
```

```python
# Sobel Vertical Filter
vertical_kernel = numpy.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
#convoluting the filter with the image
filtered_image_vertical =  signal.convolve2d(task6_image , vertical_kernel)
plt.imshow(filtered_image_vertical , cmap = "gray")
plt.show()
```

```python
# Finding the sobel magnitude
sobel_magnitude = numpy.sqrt(filtered_image_vertical**2 + filtered_image_horizontal**2)
sobel_magnitude = (sobel_magnitude - numpy.min(sobel_magnitude))/(numpy.max(sobel_magnitude) - numpy.min(sobel_magnitude)) * 255
plt.imshow(sobel_magnitude, cmap='gray')
plt.show()
```
